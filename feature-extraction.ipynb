```python
import pandas as pd
import numpy as np
import pickle
```

```python
df=pickle.load(open('dataset_level2.pkl','rb')) #we have stored the dataset in a variable and opened in read binary mode
```

```python
df
```

```python
#now we need the following features
#batting_team
#bowling_team
#city
#current score
#ball left
#wickets left
#current rr
#last five ------------>8 features
```

```python
#now checking whether every rows has not null values or not and if null values are found we will remove it
```

```python
df.isnull().sum() #here  city has many null value so to fix this we will do the below steps
```

```python
#extracting the stadium names and finding the how many time that stadium is present in a particular row
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities=np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city']) #we have extracted the first word
```

```python
df['city']=cities #after extraxtion,we have filled where the cities values are empty
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True) #droping the venue column from the dataset
```

```python
df
```

```python
#now for cities we are taking those cities value minimum 600 balls i.e atleast 5 matches has been played but not less than 5 
```

```python
eligible_cities=df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist() #extracted the cities and converted them to list
```

```python
#now removing the cities where less than 600 balls where bowled 
```

```python
df=df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
#now fetching the runs in every over and putted in a new column name current_score
df['current_score']=df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
#splitted the ball col into 2 parts over and ball no
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
#no finding how many bowls bowled till now and multiplied it by 6
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x) #here if more than 120 balls have been bowled(cause of extra) then take it as 0
df                           
```

```python
#now finding after how many balls,how many wickets are left
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
df
```

```python
#now finding the current run-rate
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
#now finding to the last 5 overs score
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five #adding a new column and adding last_five attribute there
```

```python
df #fist 30 over value is giving NAN(Not A Number) because for gettig the last 5 overs value minimum 30 balls needs to be played
```

```python
#finding total how many runs have been scored and on which ball how much run have been scored
```

```python
 final_df=df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
#now extracting the columns that we have told at the first
#df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','']] ->features
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
#now removing the missing values from last_five attribute
```

```python
final_df.dropna(inplace=True) 
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0]) #for doing suffling so that no biased gets developed
```

```python
final_df   #runs_x is our output
```

```python
final_df.sample(2)
```

```python
#now extracting x-train,y-train,x-test,y-test for model development
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor         
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(handle_unknown='ignore',drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe
```

```python
#here we have used XG-Boost algorithm
```

```python
#training the model and then fiding the r2 score and mean error
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
